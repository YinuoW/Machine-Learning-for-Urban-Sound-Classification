```python
import pandas as pd
import numpy as np
from fastai.vision import *
```

```python
path = Path('')
np.random.seed(42)
data = ImageDataBunch.from_csv(path,csv_labels='data8k/data_test10_train.csv', folder="data8k/melimg", valid_pct=0, suffix='.jpg',
        ds_tfms=get_transforms(), size=224, num_workers=0).normalize(imagenet_stats)
```

```python
data.classes
```

```python
learn = cnn_learner(data, models.resnet34, metrics=accuracy)
```

```python
#learn.load('stage-3')
learn.fit_one_cycle(5)
learn.unfreeze()
learn.fit_one_cycle(5)
test_csv = pd.read_csv('data8k/data_test10_test.csv') 
```

```python
test_csv.head()
```

```python
with open('output.csv',"w") as file:
    file.write("ID,Prediction\n")
    for test in test_csv.ID:
        img = open_image('data8k/melimg/'+str(test)+'.jpg')
        prediction = str(learn.predict(img)[0]).split()[0]
        #print(str(learn.predict(img)[0]))
        file.write(str(test)+','+prediction)
        file.write('\n')
```
